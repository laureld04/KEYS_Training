```python
# Libraries needed
from skimage import data
from skimage.filters import threshold_mean #import selected thresholding method
```

```python
thresh = threshold_mean(r) #establish and register thresholding method with image  
print(thresh) #threshold value 
binary = r > thresh #applys threshold to image to create binary image.
```

```python
#Code for plots and shapping
fig, axes = plt.subplots(ncols=3, figsize=(12, 2.5))
ax = axes.ravel()

ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3, sharex=ax[0], sharey=ax[0])

#Displaying first image
ax[0].imshow(r, cmap=plt.cm.gray)
ax[0].set_title('Original')
ax[0].axis('off')

#Creating histogram
ax[1].hist(r.ravel(), bins=256)
ax[1].set_title('Histogram')
ax[1].axvline(thresh, color='r')

#Displaying binary image
ax[2].imshow(binary, cmap=plt.cm.gray)
ax[2].set_title('Thresholded')
ax[2].axis('off')

plt.show()
```

```python
#Display all thresholding methods in skiimage library package
from skimage.filters import try_all_threshold 
fig, ax = try_all_threshold(r, figsize=(10, 10), verbose=False)
plt.show()
```

### Why did minimun error out?
### There is nothing wrong with the code...

```python
from skimage.filters import threshold_minimum 

img = data.page()

thresh = threshold_minimum(img) #establish and register thresholding method with image  
binary = img > thresh
fig, axes = plt.subplots(ncols=3, figsize=(12, 2.5))
ax = axes.ravel()

ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3, sharex=ax[0], sharey=ax[0])

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].hist(r.ravel(), bins=256)
ax[1].set_title('Histogram')
ax[1].axvline(thresh, color='r')


ax[2].imshow(binary, cmap=plt.cm.gray)
ax[2].set_title('Thresholded')
ax[2].axis('off')

plt.show()
```

The error occurs from the pairing of the image and filter itself. The minimum method creates the histogram of the grayscale image and smoothes it until only two maxima are present. The minimum between the two maxima is then taken as the threshold value. The minimum method did not work on the previous image because no two single maxima could reasonably be determined.

### Where is the Yen image?

```python
from skimage.filters import threshold_yen

img = data.page()

thresh = threshold_yen(img) #establish and register thresholding method with image  
binary = img > thresh
fig, axes = plt.subplots(ncols=3, figsize=(12, 2.5))
ax = axes.ravel()

ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3, sharex=ax[0], sharey=ax[0])

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].hist(r.ravel(), bins=256)
ax[1].set_title('Histogram')
ax[1].axvline(thresh, color='r')


ax[2].imshow(binary, cmap=plt.cm.gray)
ax[2].set_title('Thresholded')
ax[2].axis('off')

plt.show()
```

Yen worked on this image. Why?

Yen's algorithm tries several different paths between pixels in order to determine the shortest (best) path possible. This results in an upper threshold value to be calculated. All the pixels that are higher than that value are considered to be foreground. Yen's method technically worked but most of the image was calculated to be in the foreground. Resulting with a bleached out threshold.

# Image Analysis

## Counting "Cells"

```python
# Libraries needed
import skimage.color
import skimage.measure
```

```python
#Before we threshold, it is good practice to smooth the image
#This prevents unwanted removal of data due to large variation changes between background and foreground
#Without this can also cause the connected component analysis step to fail
blur = skimage.filters.gaussian(r,4) 
# The second input for this function is sigma
# A larger sigma value will result in less noise
plt.imshow(blur, cmap = "Greys")
```

## Try to increase the sigma value above. What happens?

```python
# Takes a binary image as input
thresh = threshold_mean(blur) #establish and register thresholding method with image  
print(thresh)
binary = blur > thresh #applys threshold to image to create binary image.
plot = plt.figure(1)
plt.imshow(r,cmap="Reds")
plt.title("Original Red Channel") # Gives title to the graph output
plot = plt.figure(2)
plt.imshow(binary, cmap='Greys') #plot binary of thresholded image
plt.title("Binary Image of Thresholded Red Channel")
```

## Before we "count" the cells, how does Connected Component Analysis Work? 

### https://datacarpentry.org/image-processing/09-connected-components/

```python
# Connected Component Analysis
labeled_img = skimage.measure.label(binary, connectivity=1)
#Connectivity is in units of orthogonal jumps ( you can only choose 1 or 2)
#Pixel connected to a second one and if reachable by 1 orthogonal jump from first pixel
print("dtype:",labeled_img.dtype)
print("Min:", np.min(labeled_img))
print("Max:",np.max(labeled_img))
```

The min and max represent the mininum and maximum **number of objects**. This is because our labeled image array created has a unique pixel value (either 0 or 1) and have been grouped by proximity.

Pixels directly next to each other and belong to the foreground class are considered the same object.
